```python
# Import required libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Create Decision Tree classifier object
clf = DecisionTreeClassifier(criterion='entropy', random_state=1)  # Using 'entropy' to build tree based on Information Gain

# Train Decision Tree Classifier
clf = clf.fit(X_train, y_train)

# Predict the response for test dataset
y_pred = clf.predict(X_test)

# Evaluate model
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred))

# Visualize the Decision Tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```

```python
# 1. Import libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import tree

# 2. Create dataset
data = {
    'Age': [22, 25, 47, 52, 46, 56, 27, 28, 35, 40],
    'Income': [15, 20, 90, 110, 95, 120, 25, 30, 45, 60],
    'Buy_Car': ['No', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'No', 'No', 'No', 'Yes']
}

df = pd.DataFrame(data)

# 3. Split Features and Target
X = df[['Age', 'Income']]   # Features
y = df['Buy_Car']           # Target

# 4. Build Decision Tree Model
model = DecisionTreeClassifier(criterion='entropy', random_state=0)
model.fit(X, y)

# 5. Visualize the Decision Tree
plt.figure(figsize=(12,8))
tree.plot_tree(model, feature_names=['Age', 'Income'], class_names=['No', 'Yes'], filled=True)
plt.show()

# 6. Predict new value
# Example: predict if 30 years old with $40k income will buy a car
prediction = model.predict([[30, 40]])
print("Prediction (30 years, $40k income):", prediction[0])
```

```python
# Practical 2: Implementation of Decision Tree

# Step 1: Import Libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
from sklearn import tree

# Step 2: Load Dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 3: Split Dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Model Building
clf = DecisionTreeClassifier(criterion="entropy", random_state=42)  # using 'entropy' for Information Gain
clf.fit(X_train, y_train)

# Step 5: Prediction
y_pred = clf.predict(X_test)

# Step 6: Evaluation
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Step 7: Visualization
plt.figure(figsize=(20,10))
tree.plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```
